Use `read_csv` to load the data from the TidyTuesday GitHub repository. The project URL is can be found at [https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/](https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/).

```r
library("tidyverse")
Sys.setenv("VROOM_CONNECTION_SIZE" = 1362144)

data_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv'

wind_turbine = read_csv(data_url, col_types = cols()) |>
                            select(objectid,province_territory,project_name,total_project_capacity_mw,turbine_identifier,turbine_number_in_project,turbine_rated_capacity_k_w,rotor_diameter_m,hub_height_m,manufacturer,model,commissioning_date,latitude,longitude,notes)

wind_turbine %>% head()
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
options(repr.plot.width = 4, repr.plot.height = 3, repr.plot.res = 300)

wind_turbine_sorted = wind_turbine %>%
    mutate(province_territory_sorted = province_territory %>% fct_infreq())

plot_1 = ggplot(wind_turbine_sorted, aes(x = province_territory_sorted))

plot_1 = plot_1 + geom_bar() + coord_flip() + labs(x = "Province Territory", y = "Frequency of turbine per Province Territory")

plot_1 = plot_1 + theme(axis.title.x = element_text(size = 7), 
                        axis.title.y = element_text(size = 9))

plot_1
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_turbine_sorted_manufacturer = wind_turbine %>%
    mutate(sorted_manufacturer = manufacturer %>% 
           fct_lump(n = 7) %>% 
           fct_infreq() %>% 
           fct_relevel('Other', after= Inf) %>%
           fct_rev()
    )

plot_2 = ggplot(wind_turbine_sorted_manufacturer, aes(y = sorted_manufacturer))

plot_2 = plot_2 + geom_bar() + labs(y = "Manufacturers", x = "Frequency of turbine per manufacturer") 

plot_2 = plot_2 + theme(axis.title.y = element_text(size = 9), 
                        axis.title.x = element_text(size = 9))

plot_2
```
